#Importing

```python

import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import shapiro, levene, f_oneway
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/johncduran/datasci_6_anova/main/datasets/diabetic_data.csv')
df

```

This dataset contains information of patients with diabetes from 130 plus hospitals in the U.S. from 1999-2008

## Research question:
Does one gender get more lab procedures done than another?

## Hypothesis

The null hypothesis would be that there are no significant differences between the means of male and female groups in terms of number of lab procedures.

Alternative hypothesis: There is a significant difference between the means of male and female groups in terms of number of lab procedures.



```python
columns_to_keep = ['gender', 'num_lab_procedures']
df = df[columns_to_keep]
```

Here I am creating a new dataframe using only the two columns that I need which are 'gender' and 'num_lab_procedures'.

```python

female_data = df[df['gender'] == 'Female']['num_lab_procedures']
male_data = df[df['gender'] == 'Male']['num_lab_procedures']
```

Here I am creating two subsets of data

```python
stat_female, p_value_female = shapiro(female_data)
stat_male, p_value_male = shapiro(male_data)
```

```python
print(f"Result of the Shapiro-Wilk Test (Female): Statistic={stat_female}, p-value={p_value_female}")
print(f"Result of the Shapiro-Wilk Test (Male): Statistic={stat_male}, p-value={p_value_male}")
```

Using the Shapiro-Wilk test both the p-values are both 0.0 which is less than the significance level of 0.05. This means that the data for both males and females is not normally distributed.

```python
stat_levene, p_value_levene = levene(female_data, male_data)

print(f"Result of the Levene Test: Statistic={stat_levene}, p-value={p_value_levene}")
```

Given the results of the Levene test where the p-value is 0.0, it can be concluded that the variances in the number of lab procedures are not equal in the male and female groups.

```python
stat_anova, p_value_anova = f_oneway(female_data, male_data)

print(f"Result of the One-way ANOVA: Statistic={stat_anova}, p-value={p_value_anova}")
```

Since the p-value in the ANOVA test is 0.33 which is greater than 0.05 and so we would not reject the null hypothesis. This means that there isn't enough evidence to say that there are significant differences in the means of the number of lab procedures in the male and female groups.
